# Load the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('../data/raw/framingham.csv')
data.head()
```

# Summary statistics

```python
data.describe()
```

# Missing Values

```python
data.isnull().sum()
```

# Distribution of Target Variable

```python
sns.countplot(x='TenYearCHD', data=data)
plt.title('Distribution of Ten-Year CHD Risk')
plt.savefig('../reports/plots/target_distribution.png')
plt.show()
```

# Feature distribution

```python
data.hist(figsize=(15, 10), bins=20)
plt.savefig('../reports/plots/EDA/numerical_distributions.png')
plt.show()
```

# Correlation matrix

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('../reports/plots/EDA/correlation_matrix.png')
plt.show()
```
